```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
ratings = pd.read_csv("/content/ratings.csv")
```

```python
ratings.head()
```

```python
ratings.info()
```

```python
ratings.shape
```

```python
movie_features = ratings.groupby('movieId').agg(
    avg_rating=('rating','mean'),
    num_reviews=('rating','count')
)
```

```python
movie_features = movie_features.reset_index()
```

```python
movie_features.head()
```

```python
# remove very rare movies
movie_features = movie_features[movie_features['num_reviews'] >= 20]
```

```python
# feature matrix
X = movie_features[['avg_rating','num_reviews']]
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    movie_features['avg_rating'],
    movie_features['num_reviews']
)

plt.xlabel("Average Rating")
plt.ylabel("Number of Reviews")
plt.title("Movie Ratings vs Popularity")

plt.show()
```

```python
movie_features[['avg_rating','num_reviews']].hist(figsize=(10,5))

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.heatmap(
    movie_features[['avg_rating','num_reviews']].corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Feature Correlation")

plt.show()
```

```python
# feature scaling
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

```

```python
# k means clustering
kmeans = KMeans(n_clusters=4, random_state=42)

movie_features['cluster'] = kmeans.fit_predict(X_scaled)
```

```python
# pca visualization
pca = PCA(n_components=2)

pca_features = pca.fit_transform(X_scaled)

plt.figure(figsize=(6,4))

sns.heatmap(
    movie_features[['avg_rating','num_reviews']].corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Feature Correlation")

plt.show()
```

```python
# plot clusters
# Add PCA components to the movie_features DataFrame
movie_features['pca1'] = pca_features[:, 0]
movie_features['pca2'] = pca_features[:, 1]

plt.figure(figsize=(8,6))

sns.scatterplot(
    x=movie_features['pca1'],
    y=movie_features['pca2'],
    hue=movie_features['cluster'],
    palette='viridis'
)

plt.title("Movie Clusters Based on Reviews")
plt.show()
```

```python
# Hierarchical clustering
hc = AgglomerativeClustering(n_clusters=4)

movie_features['hc_cluster'] = hc.fit_predict(X_scaled)
```

```python
# cluster interpretation
movie_features.groupby('cluster').mean()
```

```python
# Elbow method
from sklearn.cluster import KMeans

inertia = []

K = range(1,10)

for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(K, inertia)
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Method")

plt.show()
```

This project applied unsupervised machine learning techniques to cluster movies based on user review patterns using the ratings_small dataset. The key features used for clustering were average rating (quality of the movie) and number of reviews (popularity of the movie).

First, the dataset was preprocessed and transformed to obtain movie-level features, including the average rating and the total number of reviews for each movie. These features were scaled to ensure fair comparison during clustering.

Two clustering algorithms were applied:

-KMeans Clustering

-Hierarchical (Agglomerative) Clustering

The clustering results revealed that movies naturally group into different categories based on user review behaviour. For example:

Highly rated and highly reviewed movies formed a cluster representing popular blockbuster films.

Highly rated but less reviewed movies formed another cluster, representing niche or underrated films.

Moderately rated movies with average review counts formed a middle cluster.

Low rated and low popularity movies formed another cluster.

Visualization techniques such as scatter plots and PCA helped clearly display how movies were distributed across clusters.
